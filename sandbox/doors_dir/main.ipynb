```python
import random
import string

def generate_random_string(length):
    """Generate a random string of given length"""
    return ''.join(random.choice(string.ascii_letters) for _ in range(length))

def generate_random_content():
    """Generate a content with random words and numbers between 100 to 1000 words"""
    word_count = random.randint(400, 1000)
    content = []
    
    for _ in range(word_count):
        # Randomly decide whether to append a word or a number
        if random.choice([True, False]):
            # Append a random word (3-10 letters)
            content.append(generate_random_string(random.randint(3, 10)))
        else:
            # Append a random number
            content.append(str(random.randint(0, 10000)))

    return ' '.join(content)

for i in range(1, 99):
    random_string = generate_random_string(random.randint(3, 8))
    filename = f"{i}_{random_string}.txt"
    
    with open(filename, 'w') as file:
        file.write(generate_random_content())
```

```python
import csv
import random
from datetime import datetime, timedelta

def generate_random_date(start, end):
    """Generate a random date between the given start and end dates."""
    return start + timedelta(
        seconds=random.randint(0, int((end - start).total_seconds())))

def generate_random_string(length):
    """Generate a random string of given length."""
    return ''.join(random.choice(string.ascii_letters) for _ in range(length))

start_date = datetime(2000, 1, 1)
end_date = datetime(2023, 1, 1)

for i in range(1, 31):
    # Decide the number of rows in the CSV (for this example, between 10 and 500)
    num_rows = random.randint(10, 500)
    
    filename = f"random_data_{i}.csv"

    with open(filename, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['Date', 'Number', 'String'])  # Writing header

        for _ in range(num_rows):
            random_date = generate_random_date(start_date, end_date).strftime('%Y-%m-%d')
            random_number = random.randint(0, 10000)
            random_str = generate_random_string(random.randint(3, 10))

            csv_writer.writerow([random_date, random_number, random_str])

    print(f"Generated {filename} with {num_rows} rows.")
```

```python
import os
import random

class RandomABIntUpdater:
    def __init__(self):
        self.a = 0
        self.b = 0
        self.a_new = 0
        self.b_new = 0
    
    def update(self):
        while self.a_new == self.a and self.b_new == self.b:
            self.a_new = random.randint(1, 9)
            self.b_new = random.randint(1, 9)
            sum_result = self.a_new + self.b_new
        
        self.a, self.b = self.a_new, self.b_new
        
        return sum_result, self.a, self.b


updater = RandomABIntUpdater()

def main():
    response = "run"
    score = 0
    sort_function = "Size"

    while response != "q":
        correct, a, b = updater.update()

        os.system('clear')
        print(f"score = {score}\n")
        print(f"What is the sum of\n\n{a} + {b} = ?\n")
        print("quit = q\n")

        response = input("> ")
        
        if len(response) == 1 and not response.isdigit():
            os.system('clear')
            print("Look under the table")
            break

        if response == "q":
            os.system('clear')
            print(f"Your score is {score}")
            break

        try:
            if int(response) == correct:
                os.system('clear')
                print("correct")
                score += 1
            else:
                os.system('clear')
                print("try again")
        except ValueError:  
            os.system('clear')
            print(f"Sort by {sort_function}")
            break

if __name__ == "__main__":
    main()
```
