```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/Programme_for_International_Student_Assessment#:~:text=The%20results%20of%20PISA%202018,top%20performer%20in%20all%20categories.'

# Use the read_html function to scrape tables
tables = pd.read_html(url)

# If there's only one table, you can access it with tables[0]
# If there are multiple tables, iterate over the tables list
for i, table in enumerate(tables):
    print(f"Table {i}")
    print(table)
    print("\n" + "-"*50 + "\n")
```
